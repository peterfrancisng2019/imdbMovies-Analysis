## Business Question 12 -
Classify movies based on ratings (Excellent, Good and Average)



```python
# Let's create a user-defined function to help us categorize the ratings into Excellent, Good and Average respectively

def movie_rating(rating) :
    if rating>=7.0 :
        return "Excellent"
    elif rating >=6.0 :
        return "Good"
    else :
        return "Average"

```

```python
# Now let's go ahead and add a new column - "Movies_Cat" using our user-defined function
#moviesdb["Movie_Cat"] = moviesdb["Rating"].apply(movie_rating)
moviesdb["Movie_Cat"] = moviesdb["Rating"].apply(lambda x: movie_rating(x))
```

```python
moviesdb.head()
```

```python
#Classify movies based on ratings (Excellent, Good and Average)

movcat = moviesdb.groupby("Movie_Cat")[["Title","Genre"]].count()
movcat
```

```python
sns.barplot(x=movcat.index, y=movcat["Title"], data=movcat)
plt.title("Movies based on Category")
plt.show()
```
